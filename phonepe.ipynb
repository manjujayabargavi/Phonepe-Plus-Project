```python
import os
import json
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
#aggregate_transaction

agg_trans_path="C:/Users/Happy/Desktop/capstone/phonepe/pulse/data/aggregated/transaction/country/india/state/"
aggregation_list= os.listdir(agg_trans_path)

Agg_tra = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}
for i in aggregation_list:
    year = agg_trans_path+i+"/"
    year_list = os.listdir(year)

    for j in year_list:
        file= year + j +"/"
        file_list= os.listdir(file)
        
        for k in file_list:
            folder = file+k
            data= open(folder,"r")
            first=json.load(data)

            for l in first["data"]["transactionData"]:
                name=l["name"]
                count = l['paymentInstruments'][0]['count']
                amount = l['paymentInstruments'][0]['amount']
                Agg_tra['State'].append(i.title().replace('-',' '))
                Agg_tra['Year'].append(j)
                Agg_tra['Quarter'].append(int(k.strip('.json')))
                Agg_tra['Transaction_type'].append(name)
                Agg_tra['Transaction_count'].append(count)
                Agg_tra['Transaction_amount'].append(amount)
agg_trans_df=pd.DataFrame(Agg_tra)
#agg_trans_df.to_csv('aggregated_transactions',index=False)
```

```python
#aggregate user
agg_user_path =("C:/Users/Happy/Desktop/capstone/phonepe/pulse/data/aggregated/user/country/india/state/")
agg_user_list = os.listdir(agg_user_path)

agg_user = {'State':[], 'Year':[], 'Quater':[], 'Brand':[], 'User_count':[], 'User_percentage':[]}

for i in agg_user_list:
    user_year= agg_user_path+i+"/"
    user_year_list=os.listdir(user_year)

    for j in user_year_list:
        user_file= user_year+j+"/"
        user_file_list= os.listdir(user_file)
        
        for k in user_file_list:
            user_folder= user_file + k
            user_data = open(user_folder,"r")
            user= json.load(user_data)

            try:

                for l in user['data']['usersByDevice']:
                        agg_user['State'].append(i.title().replace('-',' '))
                        agg_user['Year'].append(j)
                        agg_user['Quater'].append(int(k.strip('.json')))
                        brand= l['brand']
                        count=l['count']
                        percent=l['percentage']
                        agg_user['Brand'].append(brand)
                        agg_user['User_count'].append(count)
                        agg_user['User_percentage'].append(percent*100)
            except:
                pass
agg_user_df = pd.DataFrame(agg_user)   
#agg_user_df.to_csv('aggregated_user',index=False)             
```

```python
#Map transaction
map_trans_path= ("C:/Users/Happy/Desktop/capstone/phonepe/pulse/data/map/transaction/hover/country/india/state/")
map_trans_list=os.listdir(map_trans_path)

map_transaction={'State':[],'District':[],'Year':[],'Quater':[],'Transaction_count':[],'Transaction_amount':[]}

for i in map_trans_list:
    year_tran=map_trans_path+i+"/"
    year_tran_list=os.listdir(year_tran)
    
    for j in year_tran_list:
        file_tran= year_tran+j+"/"
        file_list=os.listdir(file_tran)
        
        for k in file_list:
            trans_folder=file_tran+k
            folder = open(trans_folder,"r")
            map_tran= json.load(folder)
            
            for l in map_tran['data']['hoverDataList']:
                district=l['name']
                count=l['metric'][0]['count']
                amount=l['metric'][0]['amount']
                map_transaction['State'].append(i.title().replace('-',' '))
                map_transaction['Year'].append(j)
                map_transaction['Quater'].append(int(k.removesuffix('.json')))
                map_transaction['District'].append(district.removesuffix('district').title().replace('And','and').replace('andaman','Andaman'))
                map_transaction['Transaction_count'].append(count)
                map_transaction['Transaction_amount'].append(amount)
map_tran_df= pd.DataFrame(map_transaction)
#map_tran_df.to_csv('map_transaction',index=False)
```

```python
#map user data
map_user_path= ("C:/Users/Happy/Desktop/capstone/phonepe/pulse/data/map/user/hover/country/india/state/")
map_user_list=os.listdir(map_user_path)

map_user={'State':[],'District':[],'Year':[],'Quater':[],'Registered_user':[],'Appopened':[]}

for i in map_user_list:
    year_user=map_user_path+i+"/"
    year_user_list=os.listdir(year_user)
    
    for j in year_user_list:
        file_user= year_user+j+"/"
        file_list=os.listdir(file_user)
        
        for k in file_list:
            user_folder=file_user+k
            folder = open(user_folder,"r")
            map_usr= json.load(folder)

            for l,m in map_usr['data']['hoverData'].items():
                reg_user = m['registeredUsers']
                app_open = m['appOpens']
                map_user['State'].append(i.title().replace('-',' '))
                map_user['District'].append(l.removesuffix('district').title().replace('And','and').replace('andaman','Andaman'))
                map_user['Year'].append(j)
                map_user['Quater'].append(int(k.removesuffix('.json')))
                map_user['Registered_user'].append(reg_user)
                map_user['Appopened'].append(app_open)

map_user_df=pd.DataFrame(map_user)
#map_user_df.to_csv('map_user',index=False)
```

```python
#top transactions district
top_tran_path= ("C:/Users/Happy/Desktop/capstone/phonepe/pulse/data/top/transaction/country/india/state/")
top_tran_list=os.listdir(top_tran_path)

top_trans={'State':[],'District':[],'Year':[],'Quater':[],'Transaction_count':[],'Transaction_amount':[]}

for i in top_tran_list:
    year_tran=top_tran_path+i+"/"
    year_tran_list=os.listdir(year_tran)
    
    for j in year_tran_list:
        file_tran= year_tran+j+"/"
        file_list=os.listdir(file_tran)
        
        for k in file_list:
            tran_folder=file_tran+k
            folder = open(tran_folder,"r")
            top_tran= json.load(folder)

            for l in top_tran['data']['districts']:
                district= l['entityName']
                count=l['metric']['count']
                amount=l['metric']['amount']
                top_trans['State'].append(i.title().replace('-',' '))
                top_trans['District'].append(district.title().replace('And','and').replace('andaman','Andaman'))
                top_trans['Year'].append(j)
                top_trans['Quater'].append(int(k.removesuffix('.json')))
                top_trans['Transaction_count'].append(count)
                top_trans['Transaction_amount'].append(amount)

top_trans_dist_df=pd.DataFrame(top_trans)
#top_trans_dist_df.to_csv('top_transaction_district',index=False)
```
